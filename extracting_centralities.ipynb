```python
import os
import pandas as pd

def print_header(name):
    """Function to print out the header for centrality results"""
    print("-"*80)
    print("Centrality Results for file -> '{}' <-".format(name))
    print("-"*80)
    
def print_names(series, cent_type):
    """Function to print out centrality results nicely"""
    if cent_type == 'btwn':
        print("Handles of users with highest Betweeness Centrality:")
    elif cent_type == 'close':
        print("Handles of users with highest Closeness Centrality:")
    names = series.tolist()
    for name in names:
        print("---->", name)
    print("")

#Extract gephi table files
root_path = '../data/Gephi_data/centralities/'
files = os.listdir(root_path)

#Get the centrality results
for file in files:
    print_header(file)
    df = pd.read_csv(root_path+file)
    btwn_cent = df.sort_values('betweenesscentrality', ascending=False)
    close_cent = df.sort_values('closnesscentrality', ascending=False)
    top_btwn_cent = btwn_cent.head()['Label']
    top_close_cent = close_cent.head()['Label']
    print_names(top_btwn_cent, 'btwn')
    print_names(top_close_cent, 'close')
#     break
```
